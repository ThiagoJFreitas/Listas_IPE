```python
#QUESTAO 1) Contruir um algoritmo que leia dois vetores (A e B) de 50 elementos cada e crie um terceiro
#  vetor acumulando a soma dos elementos desses dois vetores, sempre utilizando o mesmo índice.
#  Mostre os valores do terceiro vetor.
#  Exemplo: c[1]← A[1]+B[1];
# Função para ler um vetor de 50 elementos
# Função para ler um vetor de 50 elementos, aceitando letras ou números
def ler_vetor(nome):
    vetor = []
    print(f"Digite os valores do vetor {nome}:")
    for i in range(5):
        valor = input(f"{nome}[{i+1}]: ")  # Aceita qualquer entrada (letras ou números)
        try:
            # Tenta converter para número se possível
            valor = float(valor) if '.' in valor else int(valor)
        except ValueError:
            # Se não for um número, permanece como string
            pass
        vetor.append(valor)
    return vetor

# Função para criar o vetor C com a soma ou concatenação dos elementos de A e B
def criar_vetor_soma(A, B):
    C = []
    for i in range(5):
        if isinstance(A[i], (int, float)) and isinstance(B[i], (int, float)):
            # Soma se ambos forem números
            soma = A[i] + B[i]
        else:
            # Concatena se algum dos elementos for string
            soma = str(A[i]) + str(B[i])
        C.append(soma)
    return C

# Função para exibir o vetor C
def exibir_vetor(C):
    print("Vetor C (soma ou concatenação dos elementos de A e B):")
    for i in range(5):
        print(f"C[{i+1}] = {C[i]}")

# Leitura dos vetores A e B
A = ler_vetor("A")
B = ler_vetor("B")

# Criação do vetor C com a soma ou concatenação de A e B
C = criar_vetor_soma(A, B)

# Exibição do vetor C
exibir_vetor(C)

```

```python
#QUESTAO 2 -
# Função para gerar a matriz 5x5 com a diagonal principal de 0 e o restante de 1
def gerar_matriz():
    tamanho = 5
    matriz = []

    # Iterar sobre as linhas
    for i in range(tamanho):
        linha = []
        for j in range(tamanho):
            if i == j:  # Verifica se está na diagonal principal
                linha.append(0)  # Adiciona 0 na diagonal principal
            else:
                linha.append(1)  # Adiciona 1 no restante da matriz
        matriz.append(linha)

    return matriz

# Função para exibir a matriz
def exibir_matriz(matriz):
    for linha in matriz:
        print(" ".join(map(str, linha)))

# Gerar e exibir a matriz
matriz = gerar_matriz()
exibir_matriz(matriz)
```

```python
#Questão 3 Leia uma matriz 4 x 4, conte e escreva quantos valores maiores que 10 ela possui

# Função para ler uma matriz 4x4
def ler_matriz():
    matriz = []
    print("Digite os valores da matriz 4x4:")
    for i in range(4):
        linha = []
        for j in range(4):
            valor = int(input(f"Elemento [{i+1}, {j+1}]: "))  # Leitura de cada elemento
            linha.append(valor)
        matriz.append(linha)
    return matriz

# Função para contar quantos valores maiores que 10 existem na matriz
def contar_valores_maiores_que_10(matriz):
    contador = 0
    for linha in matriz:
        for valor in linha:
            if valor > 10:
                contador += 1
    return contador

# Leitura da matriz
matriz = ler_matriz()

# Contagem de valores maiores que 10
quantidade = contar_valores_maiores_que_10(matriz)

# Exibição do resultado
print(f"A matriz possui {quantidade} valores maiores que 10.")
```

```python
#Questão 4 - Preencha uma matriz 5 x 5 com números aleatórios. Leia também um valor X. O programa
# deverá fazer uma busca desse valor na matriz e, ao final, escrever a localização (linha e coluna)
#  ou uma mensagem de “não encontrado”

import random

# Função para gerar uma matriz 5x5 com números aleatórios
def gerar_matriz_aleatoria():
    matriz = []
    for i in range(5):
        linha = [random.randint(1, 100) for _ in range(5)]  # Números aleatórios entre 1 e 100
        matriz.append(linha)
    return matriz

# Função para exibir a matriz
def exibir_matriz(matriz):
    for linha in matriz:
        print(" ".join(map(str, linha)))

# Função para buscar o valor X na matriz
def buscar_valor(matriz, valor):
    for i in range(5):
        for j in range(5):
            if matriz[i][j] == valor:
                return (i, j)  # Retorna a posição (linha, coluna) se o valor for encontrado
    return None  # Retorna None se o valor não for encontrado

# Geração da matriz 5x5 com números aleatórios
matriz = gerar_matriz_aleatoria()

# Exibição da matriz
print("Matriz gerada:")
exibir_matriz(matriz)

# Leitura do valor X
x = int(input("Digite o valor a ser buscado (X): "))

# Busca do valor X na matriz
resultado = buscar_valor(matriz, x)

# Exibição do resultado
if resultado:
    linha, coluna = resultado
    print(f"Valor {x} encontrado na posição (linha {linha+1}, coluna {coluna+1}).")
else:
    print(f"Valor {x} não encontrado na matriz.")
```

```python
# Questão 05) Faça um programa para gerar automaticamente números entre 0 e 99 de uma cartela de
# bingo. Sabendo que cada cartela deverá conter 5 linhas de 5 números, gere estes dados de modo
# a não ter números repetidos dentro das cartelas. O programa deve exibir na tela a cartela gerada.

import random

# Função para gerar uma cartela de bingo 5x5 sem números repetidos
def gerar_cartela_bingo():
    numeros_disponiveis = list(range(0, 100))  # Números de 0 a 99
    random.shuffle(numeros_disponiveis)  # Embaralha os números

    cartela = []
    for i in range(5):
        linha = []
        for j in range(5):
            numero = numeros_disponiveis.pop()  # Retira o último número da lista (sem repetir)
            linha.append(numero)
        cartela.append(linha)

    return cartela

# Função para exibir a cartela de bingo
def exibir_cartela(cartela):
    print("Cartela de Bingo:")
    for linha in cartela:
        print(" ".join(f"{num:02d}" for num in linha))  # Exibe números com 2 dígitos

# Geração da cartela de bingo
cartela = gerar_cartela_bingo()

# Exibição da cartela
exibir_cartela(cartela)
```

```python
# Questão 06) Uma turma de formandos está vendendo rifas para angariar recursos financeiros para sua
#  cerimônia de formatura. Construa um programa para cadastrar os nomes das pessoas que
#  compraram a rifa. Ao fim, o programa deve sortear o ganhador do prêmio e imprimir o seu nome.

import random

# Função para cadastrar os nomes dos compradores
def cadastrar_compradores():
    compradores = []
    while True:
        nome = input("Digite o nome do comprador (ou pressione ENTER para encerrar): ")
        if nome == "":  # Se o usuário pressionar Enter sem digitar, encerra o cadastro
            break
        compradores.append(nome)
    return compradores

# Função para sortear o ganhador
def sortear_ganhador(compradores):
    if not compradores:
        print("Nenhum comprador cadastrado para o sorteio.")
        return None
    ganhador = random.choice(compradores)  # Escolhe um nome aleatório da lista
    return ganhador

# Cadastro dos compradores
compradores = cadastrar_compradores()

# Sorteio do ganhador
if compradores:
    ganhador = sortear_ganhador(compradores)
    print(f"O ganhador da rifa é: {ganhador}")
else:
    print("Não há compradores cadastrados para realizar o sorteio.")
```
